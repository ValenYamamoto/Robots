```python
import df_operations
import pandas
import seaborn as sns
from matplotlib.lines import Line2D
import matplotlib.pyplot as plt
```

```python
df1 = df_operations.read_pickle( "0214-1358-pytorch.p" )
df2 = df_operations.read_pickle( "0215-1040-pytorch.p" )
#print( df1, df2 )
data = df_operations.combine_df( df1, df2, 30 )
```

```python
p = sns.lineplot( data=data, x="epoch", y="actor loss" )
sns.lineplot( data=data, x="epoch", y="critic loss", ax=p.axes.twinx(), color='y' )
p.legend(handles=[Line2D([], [], marker='_', color="b", label='actor loss'), Line2D([], [], marker='_', color="y", label='critic loss')])
plt.savefig( "actor_critic_loss.png" )
```
