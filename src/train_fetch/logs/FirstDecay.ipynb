```python
import pandas
import seaborn as sns
from matplotlib.lines import Line2D
import matplotlib.pyplot as plt
import df_operations
import histograms
```

```python
df1 = df_operations.read_pickle( "0312-1925-pytorch.p" )
df2 = df_operations.read_pickle( "0315-1140-pytorch.p" )
df = df_operations.combine_df( df1, df2, 35 )
df
```

```python
sns.lineplot( data=df, x='epoch', y="avg rewards" )
plt.title( "Avg Reward")
```

```python
p = sns.lineplot( data=df, x='epoch', y="total rewards" )
sns.lineplot( data=df, x="epoch", y="avg entropy", ax=p.axes.twinx(), color='y' )
p.legend(handles=[Line2D([], [], marker='_', color="b", label='rewards'), Line2D([], [], marker='_', color="y", label='entropy')])
plt.title( "Total Reward")
```

```python
p = sns.lineplot( data=df, x='epoch', y="critic loss" )
sns.lineplot( data=df, x="epoch", y="hit goal", ax=p.axes.twinx(), color='y' )
p.legend(handles=[Line2D([], [], marker='_', color="b", label='loss'), Line2D([], [], marker='_', color="y", label='hit goal')])
plt.title( "Critic Loss")
```
