```python
import pandas
import seaborn as sns
from matplotlib.lines import Line2D
import matplotlib.pyplot as plt
import df_operations
```

```python
df1 = df_operations.read_pickle( "0223-1932-pytorch-action.p" )
```

```python
sns.lineplot( data=df1, x="epoch", y="action average" )
```

```python
sns.lineplot( data=df1, x="epoch", y="collision" )
```

```python
df2 = df_operations.read_pickle( "0223-1932-pytorch.p" )
p = sns.lineplot( data=df2, x="epoch", y="critic loss", color='r' )
sns.lineplot( data=df2, x="epoch", y="actor loss", ax=p.axes.twinx())
p.legend(handles=[Line2D([], [], marker='_', color="b", label='actor loss'), Line2D([], [], marker='_', color="r", label='critic loss')])
```

```python
p = sns.lineplot( data=df2, x="epoch", y="total rewards", color='r' )
sns.lineplot( data=df1, x="epoch", y="collision", ax=p.axes.twinx())
p.legend(handles=[Line2D([], [], marker='_', color="b", label='avg collisions'), Line2D([], [], marker='_', color="r", label='total rewards')])
```
